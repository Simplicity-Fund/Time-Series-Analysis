## Calculate Autocorrelation

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

from pandas.plotting import lag_plot
import matplotlib.pyplot as plt

import yfinance as yf
```

### Autocorrelation Function (ACF) Plot

The Autocorrelation Function (ACF) plot is used to visualize how values at different lags are correlated with the present value. The plot shows the autocorrelation values for a range of lags. Significant spikes in the ACF plot indicate strong correlations at those lags.

```python
ticker = "AAPL"
start_date = "2020-01-01"
end_date = "2023-12-31"

# Fetch the stock data
apple_data = yf.download(ticker, start=start_date, end=end_date)
```

```python
# Plot the ACF for the original Apple stock price series
plt.figure(figsize=(10, 5))
plot_acf(apple_data['Close'].dropna(), lags=30)
plt.title('Autocorrelation Function (ACF) - Apple Stock Prices')
plt.show()
```

## Partial Autocorrelation Function (PACF) Plot

The Partial Autocorrelation Function (PACF) plot shows the correlation between a time series and its lags after removing the influence of intermediate lags. This helps to isolate the direct correlation between a value and a specific lag.

```python
# Plot the PACF for Apple stock price series
plt.figure(figsize=(10, 5))
plot_pacf(apple_data['Close'].dropna(), lags=30)
plt.title('Partial Autocorrelation Function (PACF) - Apple Stock Prices')
plt.show()
```

## Time Series Lag Plot

A lag plot visualizes the relationship between the values of the series at time t and time t-k (for different lag values k). This plot helps you see patterns and relationships between current and past values of the series.

```python
# Create a lag plot with a lag of 1
plt.figure(figsize=(6, 6))
lag_plot(apple_data['Close'].dropna(), lag=12)
plt.title('Lag Plot - Apple Stock Prices (Lag=12)')
plt.show()
```
